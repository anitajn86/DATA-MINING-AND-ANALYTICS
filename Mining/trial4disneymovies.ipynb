```python
import requests
import csv
from bs4 import BeautifulSoup

# Your ScraperAPI key (Replace with your actual key)
API_KEY = "7a4419621f83d13d38b016f41da1d1df"

def scrape_disney_movies(movie_urls, output_csv="disney_movies_data.csv"):
    headers = {"User-Agent": "Mozilla/5.0"}
    
    # Create CSV file and write headers only once
    with open(output_csv, mode="w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)
        writer.writerow(["Title", "Rating", "Duration", "Release Date", "Genre", "Synopsis", "Director", "Cast", "Awards"])

    for movie_url in movie_urls:
        # Use ScraperAPI to bypass restrictions
        proxied_url = f"https://api.scraperapi.com/?api_key={API_KEY}&url={movie_url}"
        response = requests.get(proxied_url, headers=headers)

        if response.status_code != 200:
            print(f"Error: Unable to fetch {movie_url} (Status Code: {response.status_code})")
            continue  # Skip to the next movie if there's an error

        soup = BeautifulSoup(response.text, "html.parser")

        # Extract movie details
        title = soup.select_one('h1.catalog-title')
        title = title.text.strip() if title else "N/A"

        # Scraping meta values (Rating, Duration, Release Date, Genre)
        meta_values = [value.text.strip() for value in soup.find_all('p', class_='meta-value')]

        rating = meta_values[0] if len(meta_values) > 0 else "N/A"
        duration = meta_values[1] if len(meta_values) > 1 else "N/A"
        release_date = meta_values[2] if len(meta_values) > 2 else "N/A"
        genre = meta_values[3] if len(meta_values) > 3 else "N/A"

        # Scraping the synopsis (Shortened to 20 words)
        synopsis = soup.select_one('p')
        if synopsis:
            words = synopsis.text.strip().split()
            synopsis = " ".join(words[:20]) + "..." if len(words) > 20 else synopsis.text.strip()
        else:
            synopsis = "N/A"

        # Extracting movie credits properly
        def get_credit_value(credit_title):
            """Finds movie credits by matching titles within the 'ul.credits' section."""
            credits_section = soup.find("ul", class_="credits")
            if not credits_section:
                return "N/A"

            for item in credits_section.find_all("li"):
                title_tag = item.find("h2", class_="credits-title")
                value_tag = item.find("div", class_="credits-value")

                if title_tag and value_tag and credit_title in title_tag.text:
                    return value_tag.text.strip()

            return "N/A"

        director = get_credit_value("Directed By")
        cast = get_credit_value("Cast")
        awards = get_credit_value("Awards")

        # Append to CSV file
        with open(output_csv, mode="a", newline="", encoding="utf-8") as file:
            writer = csv.writer(file)
            writer.writerow([title, rating, duration, release_date, genre, synopsis, director, cast, awards])

        print(f"✅ Successfully scraped: {title}")


```
