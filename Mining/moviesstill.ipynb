```python
import requests
import csv
from bs4 import BeautifulSoup

API_KEY = "6cd695ab"  # Replace with your OMDb API key
BASE_URL = "http://www.omdbapi.com/"

def get_imdb_titles():
    url = "https://www.imdb.com/chart/moviemeter/?ref_=nv_mv_mpm"  # IMDb "Top Rated English Movies"
    headers = {"User-Agent": "Mozilla/5.0"}  # To avoid bot detection
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        movies = soup.select(".ipc-title__text")  # Find movie title elements
        movie_titles = [movie.get_text() for movie in movies]  # Get all movies
        return movie_titles
    else:
        print("âŒ Error fetching IMDb page.")
        return []

def search_movies_and_save(titles, filename="movies.csv"):
    with open(filename, mode="w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)
        writer.writerow(["Title", "Year", "IMDb Rating", "Genre", "Director", "Plot"])

        for title in titles:
            params = {"apikey": API_KEY, "t": title}
            response = requests.get(BASE_URL, params=params)

            if response.status_code == 200:
                data = response.json()
                if data["Response"] == "True":
                    writer.writerow([
                        data.get("Title", "N/A"),
                        data.get("Year", "N/A"),
                        data.get("imdbRating", "N/A"),
                        data.get("Genre", "N/A"),
                        data.get("Director", "N/A"),
                        data.get("Plot", "N/A")
                    ])
                    print(f"âœ… Saved: {data['Title']}")
                else:
                    print(f"âŒ Movie '{title}' not found.")
            else:
                print(f"âš ï¸ Error fetching data for '{title}'.")

# Scrape ALL movie titles from IMDb's "Top Rated English Movies"
titles = get_imdb_titles()

# Fetch movie details and save to CSV
search_movies_and_save(titles)

print("\nğŸ“ Data saved to 'movies.csv' successfully! ğŸš€")
```
