```python
import pandas as pd
```

```python
studentdata=pd.read_excel('students.xlsx')
studentdata
```

```python
#checking for missing values
missingvalues=studentdata.isna().sum()
missingvalues
```

```python
# Imputation with mode for 'COURSE' and 'GENDER'
studentdata[['COURSE', 'GENDER']] = studentdata[['COURSE', 'GENDER']].fillna(studentdata[['COURSE', 'GENDER']].mode().iloc[0])
studentdata
```

```python
#checking for missing values
missingvalues=studentdata.isna().sum()
missingvalues
```
