```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data['Date']=data['Date'].astype('datetime64[ns]')
```

```python
data.head(3)
```

```python
data['Motiv']=data['Date'].dt.month
```

```python
data.head(3)
```

```python
data['Country']=data['Country'].astype(str)
```

```python
data['CutCountry']=data['Country'].str[:3]
```

```python
data.tail(5)
```

```python
data['Profit&Cost']=data['Profit'].fillna(0) + data['Cost'].fillna(0)
```

```python
data['Rev-Profit']=data['Revenue']-data['Profit']
```

```python
data.head(5)
```

```python
#univariate analysis
#continuous histogram
plt.figure(figsize=(12,8))
sns.histplot(data=data,x='Revenue',bins=20,kde=True,color='green')
plt.title('Hostogram of revenue')
plt.show()
```

```python
#boxplot
plt.figure(figsize=(12,8))
sns.boxplot(data=data,y='Profit',color='green')
plt.title('Boxplot of profit')
plt.show()
```

```python
#countplot for categorical data
plt.figure(figsize=(12,8))
sns.countplot(data=data,x='Country',palette='viridis')
plt.title('Seeing the different countries')
plt.show()
```

```python
#cont vs cont
plt.figure(figsize=(12,8))
sns.scatterplot(data=data,x='Cost',y='Revenue',alpha=0.6,color='green')
plt.title('Does revenue increase when cost increases?')
plt.xlabel('Cost')
plt.ylabel('Revenue')
plt.show()
```

```python
#cont vs cat
plt.figure(figsize=(12,8))
sns.boxplot(data=data,x='Country',y='Revenue',palette='viridis')
plt.title('Revenue in different countries')
plt.xlabel('Countries')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
data['Year'].dtype
```

```python
data['Year']=data['Year'].astype('int')
```

```python
rev_years=data.groupby('Year')['Revenue'].sum()
rev_years
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(data=data,x='Year',y='Revenue',color='green')
plt.title('Revenue over the years')
plt.xlabel('Years')
plt.ylabel('Revenue')
plt.show()
```

```python
#cat vs cat
pivoted=pd.crosstab(data['Country'], data['State'])
pivoted
```

```python
pivoted.plot(kind='bar',stacked=True,figsize=(12,8),colormap='viridis')
plt.title('State by country')
plt.xlabel('State')
plt.ylabel('Country')
plt.show()
```

```python
#cat vs cat
pivoted1=pd.crosstab(data['Country'], data['Age_Group'])
pivoted1
```

```python
pivoted1.plot(kind='bar',stacked=True,figsize=(12,8),colormap='viridis')
plt.title('Age_Group by country')
plt.xlabel('Country')
plt.ylabel('Age_Group')
plt.show()
```

```python
#multivariate
plt.figure(figsize=(12,8))
sns.heatmap(data[['Revenue','Cost','Profit','Year']].corr(),annot=True,cmap='viridis',fmt='.2f')
plt.title('Heatmap of the data')
plt.show()
```
